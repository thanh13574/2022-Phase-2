# Data Modelling

```python
# An exhaustive list of models can be found in sci-kit learn in this link: https://scikit-learn.org/stable/supervised_learning.html
# The url takes you to supervised learning which is what you will be dealing with for this phase.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

dataset = pd.read_csv("weather-data.csv")
dataset.drop(columns = [" station","valid","dwpc","relh","sknt","gust","peak_wind_drct"], inplace=True) #leave only temperature in C

dataset["tmpf"]= dataset.apply((lambda x: (x["tmpc"] * 9 / 5) + 32 ), axis = 1)   #formula for conversion from C to F

X_train, X_test, y_train, y_test = train_test_split(dataset[["tmpc"]], dataset["tmpf"], test_size=0.4)  #train model

X_train, y_train = make_regression(n_samples=103713, n_features=10, noise=100, random_state=42)

trainX, testX, trainY, testY = train_test_split(X_train, y_train, test_size=0.2)

model = LinearRegression()
model.fit(trainX, trainY)
model.score(testX, testY)

model.predict(testX)
```
